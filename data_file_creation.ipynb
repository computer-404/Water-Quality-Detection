```python
import pandas as pd

train_df = pd.DataFrame()
test_df = pd.DataFrame()
```

```python
import os

directory = "train/Clean-samples"
file_paths = []
labels = []

for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        file_paths.append(os.path.join(directory, filename))
        labels.append(0)
        
directory = "train/Dirty-samples"
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        file_paths.append(os.path.join(directory, filename))
        labels.append(1)


train_df['file_paths'] = file_paths
train_df['labels'] = labels

train_df
```

```python
import os

directory = "test/Clean-samples"
file_paths = []
labels = []

for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        file_paths.append(os.path.join(directory, filename))
        labels.append(0)
        
directory = "test/Dirty-samples"
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        file_paths.append(os.path.join(directory, filename))
        labels.append(1)


test_df['file_paths'] = file_paths
test_df['labels'] = labels

test_df
```

```python
train_df.to_csv('train.csv', index=False)
test_df.to_csv('test.csv', index=False)
```
